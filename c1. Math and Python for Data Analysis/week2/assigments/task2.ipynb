```python
import math
from math import sin, cos, exp
from matplotlib import pylab as plt

import numpy as np
```

```python
def f(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
```

```python
def approx_func_values(x, coefs=[]):
    coefs = np.array(coefs)
    x_help = np.array([x ** i for i in range(0, len(coefs))])
    return coefs.dot(x_help)
```

```python
def plot_approx(x=[], coefs=[]):
    y1 = np.vectorize(f)
    y2 = [approx_func_values(x_i, coefs) for x_i in x]
    plt.plot(x, y1(x))
    plt.plot(x, y2)
    plt.show()
```

```python
def solve_lin_system(x_values=[]):
    size = len(x_values)
    A = np.zeros((size, size))
    b = np.empty(size, dtype=float)
    for i in range(0, size):
        for j in range(0, size):
            A[i, j] = x_values[i] ** j
        b[i] = f(x_values[i])
    return np.linalg.solve(A, b)
```

```python
x = np.arange(1, 15, 0.1)
```

```python
x_res = solve_lin_system([1, 15])
```

```python
plot_approx(x, x_res)
```

```python
x_res = solve_lin_system([1, 8, 15])
```

```python
plot_approx(x, x_res)
```

```python
x_res = solve_lin_system([1, 4, 10, 15])
```
